```python
#df_plot = df_years[df_years['COUNTRY']=='UKR'][df_years['Variable']=='Human consumption per capita'][df_years['Commodity']=='Fish'][['Time', 'Value']].groupby('Time').sum()

def plot_countries_consump_seperateplot(countries, commodity):
    all_df =[]
    for i in range(len(countries)):
        df_plot = df_years[df_years['COUNTRY']==countries[i]][df_years['Variable']=='Human consumption per capita'][df_years['Commodity']==commodity][['Time', 'Value']].groupby('Time').sum()
        all_df.append(df_plot)
        
    fig, axes = plt.subplots(nrows=2, ncols=2)
    all_df[0].plot(ax=axes[0,0])
    all_df[1].plot(ax=axes[0,1])
        
#plot_countries_consump_seperateplot(['ZMB'], 'Fish')


def plot_one_country_several_products_consump_oneplot(countries, commodity):
    all_df =[]
    for i in range(len(commodity)):
        df_plot = df_years[df_years['COUNTRY']==countries][df_years['Variable']=='Human consumption per capita'][df_years['Commodity']==commodity[i]][['Time', 'Value']]      
        all_df.append(df_plot)
    
    for i in range(len(all_df)):
        #print(i)
        #print(commodity[i])
        plt.plot(all_df[i].groupby('Time').sum(), label=commodity[i])
    plt.legend()
        
        
        
def plot_several_countries_one_product_consump_oneplot(countries, commodity):
    all_df =[]
    for i in range(len(countries)):
        df_plot = df_years[df_years['COUNTRY']==countries[i]][df_years['Variable']=='Human consumption per capita'][df_years['Commodity']==commodity][['Time', 'Value']]
        all_df.append(df_plot)
    
    for i in range(len(all_df)):
        #print(i)
        #print(countries[i])
        plt.plot(all_df[i].groupby('Time').sum(), label=countries[i])
        plt.legend()

        
plot_one_country_several_products_consump_oneplot('CHN', ['Fish', 'Rice', 'Wheat', 'Maize', 'Other coarse grains', 'Rice',
       "Distiller's dry grains", 'Soybean', 'Other oilseeds',
       'Protein meals', 'Vegetable oils', 'Sugar (tq)',
       'Beef and veal (cwe)', 'Pigmeat (cwe)'])
```
